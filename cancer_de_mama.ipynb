```python
# Importando bibliotécas para o modelo de Machine Learning:

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Separando as variáveis independentes da variável alvo:

X = df.drop(['diagnosis', 'id'], axis = 1)
y = df['diagnosis']
```

```python
# Padronizando as colunas númericas:

X = StandardScaler().fit_transform(X)
```

```python
# Label enconder na variável alvo:

y = LabelEncoder().fit_transform(y)
```

```python
# Dividindo os dados entre treino e teste:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# Obs: StandardScaler para calcular os scores é z = \frac{x-u}{s},
# onde u é a média das amostras do conjunto de treino e s é o desvio padrão dessa amostra.
```

```python
# Random Forest (Tentar classificar de forma correta entre um tumor maligno e um tumor benigno.)

from sklearn.ensemble import RandomForestClassifier
```

```python
# Instanciando o modelo de Random Forest:

ml_model = RandomForestClassifier(n_estimators = 10, criterion = 'entropy',
                                 random_state = 42)

# Treinando o modelo:

ml_model.fit(X_train, y_train)
```

```python
# Testando o desempenho do modelo de detecção:

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Realizando as previsões nos dados de teste:
y_pred = ml_model.predict(X_test)
```

```python
# Verificando a acurácia geral:

print('[Acurácia] Random Forest:', accuracy_score(y_test, y_pred))
```

```python
# Imprimindo o classification report:

print('\n[Classification Report] Random Forest')
print( classification_report(y_test, y_pred))
```

```python
# Plotando a matriz de confusão:

pd.DataFrame(confusion_matrix(y_test, y_pred),
            index = ['neg', 'pos'], columns = ['pred_neg', 'pred_pos'])
```
